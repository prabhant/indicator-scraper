```python
import pandas as pd
import urllib.request as urllib2
from bs4 import BeautifulSoup
import pandas as pd
```

```python
index_page = 'http://indicators.report/indicators/'
page_main = urllib2.urlopen(index_page)
soup_url = BeautifulSoup(page_main,'html.parser')
arr_url = []
for item in soup_url.find_all(attrs={'class': 'indicators-archive'}):
    for link in item.find_all('a'):
        arr_url.append(link.get('href'))
print(arr_url)
```

```python
arr_main = []
arr_name = []
for i in range(len(arr_url)):
    quote_page = arr_url[i]
    page = urllib2.urlopen(quote_page)
    soup = BeautifulSoup(page, 'html.parser')
    name_box = soup.find('h1', attrs={'class': 'entry-title'})
    name = name_box.text.strip()
    name = name[13:len(name)]
    arr_name.append(name)
    content_box = soup.find('div', attrs={'class': 'entry-content'})
    #print(content_box)
    content = content_box.text.strip()
    arr_main.append(content)
df = pd.DataFrame(arr_main)
```

```python
arr_name
```

```python
arr_big = []
for i in range(len(arr_main)):
    text = arr_main[i]
    text = text.replace('Rationale and definition:\n', '**')
    text = text.replace('\nDisaggregation:\n', '**')
    text = text.replace('\nComments and limitations:\n', '**')
    text = text.replace('\nPreliminary assessment of current data availability by Friends of the Chair:\n', '**')
    text = text.replace('\nPrimary data source:\n', '**')
    text = text.replace('\nPotential lead agency or agencies:\n', '**')
    text = text.replace('\nFootnotes\n', '**')
    text = text.split('**')
    arr_big.append(text)
    
```

```python
df = pd.DataFrame(arr_big)
df.columns = ['Indicator','Definition', 'Disaggregation','Comments',
              'Preliminary assessment of current data availability by Friends of the Chair','Indicator Source'
             ,'Potential lead agency', 'Footnotes']
df['Indicator'] = arr_name
```

```python
df.to_csv('indicator100.csv')
```
